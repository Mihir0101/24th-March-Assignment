- > 1. By deleting the row or columns.

- > 2. By Mean,Midian or Mode.

## 1.By deleting the row or columns.

Advantage - This is smallest technique to remove missing values.We can simply remove the NaN value by one function that is dropna().

Advantage -  We can handle numeric missing value by Mean&Median and for categorical missing value we can use Mode.

## Data Encoding

- > Data Encoding on Gender.

```python
df['gender'] = np.where(df['gender'].str.contains('female'),0,1)
```

```python
df.head()
```

- > Data Encoding on Race_Ethnicity

```python
# df.loc[row_value, column_value] = new_value

for i in range(1000):
    if df.loc[i,'race_ethnicity'] == 'group A':
        df.loc[i,'race_ethnicity'] = 0
    elif df.loc[i,'race_ethnicity'] == 'group B':
        df.loc[i,'race_ethnicity'] = 1
    elif df.loc[i,'race_ethnicity'] == 'group C':
        df.loc[i,'race_ethnicity'] = 2
    elif df.loc[i,'race_ethnicity'] == 'group D':
        df.loc[i,'race_ethnicity'] = 3
    elif df.loc[i,'race_ethnicity'] == 'group E':
        df.loc[i,'race_ethnicity'] = 4
```

```python
df.race_ethnicity.unique()
```

- > Data Encoding on Parental level of Education

```python
df.parental_level_of_education.unique()
```

```python
for i in range(1000):
    if df.loc[i,'parental_level_of_education'] == 'some high school':
        df.loc[i,'parental_level_of_education'] = 0
    elif df.loc[i,'parental_level_of_education'] == 'high school':
        df.loc[i,'parental_level_of_education'] = 1
    elif df.loc[i,'parental_level_of_education'] == 'some college':
        df.loc[i,'parental_level_of_education'] = 2
    elif df.loc[i,'parental_level_of_education'] == "bachelor's degree":
        df.loc[i,'parental_level_of_education'] = 3
    elif df.loc[i,'parental_level_of_education'] == "associate's degree":
        df.loc[i,'parental_level_of_education'] = 4
    elif df.loc[i,'parental_level_of_education'] == "master's degree":
        df.loc[i,'parental_level_of_education'] = 5
```

```python
df.head()
```

Q5. Load the wine quality data set and perform exploratory data analysis (EDA) to identify the distribution
of each feature. Which feature(s) exhibit non-normality, and what transformations could be applied to
these features to improve normality?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('winequality-red.csv')
```

```python
df.head()
```

```python
for i in df.columns :
    sns.histplot(x=i,data=df,kde=True)
    plt.show()
```

* All features exihibit non-normality axcept density and pH.

* These are the techniques which we can use for transformation.

1. Log Transformation

2. Square-Root Transformation

3. Reciprocal Transformation

4. Box-Cox Transformation

5. Yeo-Johnson Transformation

```python
## Features that distributed non-normaly.

non_normal = ['fixed acidity','volatile acidity','citric acid','residual sugar','chlorides','free sulfur dioxide','total sulfur dioxide','sulphates','alcohol']
```

```python
for i in non_normal :
    df[i] = np.log1p(df[i])
```

```python
df['log_target'] = np.log1p(df['fixed acidity'])
```
